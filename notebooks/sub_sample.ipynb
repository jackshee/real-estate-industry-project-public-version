```python
# merge the property_summary with the data/processed/domain/rental_listings_summary_cleaned.csv on PropertyID and property_id

property_summary = pd.read_csv('../data/curated/rent_features/property_summary.csv')
rental_listings_summary_cleaned = pd.read_csv('../data/processed/domain/rental_listings_summary_cleaned.csv')
merged = pd.merge(rental_listings_summary_cleaned, property_summary, left_on='property_id', right_on='PropertyID', how='outer').drop(columns=['property_id'])
# drop rows that come from property_summary that do not have a match in rental_listings_summary_cleaned
merged = merged[~merged['PropertyID'].isna()]
# print the number of rows in the merged dataframe
print(f'Number of rows in the merged dataframe: {len(merged)}')
# save the merged dataframe to a csv file in data/curated/rent_features/property_summary_merged.csv
merged.to_csv('../data/curated/rent_features/property_summary_merged.csv', index=False)
```

```python
# remaining rows of the rental_listings_summary_cleaned that do not have a match in the property_summary dataframe
remaining = rental_listings_summary_cleaned[~rental_listings_summary_cleaned['property_id'].isin(property_summary['PropertyID'])]
print(f'Number of remaining rows: {len(remaining)}')    
# save the remaining rows to a csv file in data/curated/rent_features/remaining_listings.csv
remaining.to_csv('../data/curated/rent_features/remaining_listings.csv', index=False)
```

```python
# code to merge isochrones for walking and driving with the listings cleaned sampled csv
listings_df_full = pd.read_csv("../data/curated/rent_features/cleaned_listings.csv")
listings_copy_df = listings_df_full.copy()

# save to a new file
listings_copy_df.to_csv("../data/curated/rent_features/cleaned_listings_venura.csv", index=False)

cols_to_add = ['5min', '10min', '15min']

driving_df_full = pd.read_csv("../isochrones/cleaned_isochrones_driving.csv")
driving_df_full = driving_df_full[["property_id"] + cols_to_add]
prefix_1 = 'driving_'
driving_df_full = driving_df_full.rename(columns={col: prefix_1 + col for col in cols_to_add})

walking_df_full = pd.read_csv("../isochrones/cleaned_isochrones_walking.csv")
walking_df_full = walking_df_full[["property_id"] + cols_to_add]
prefix_2 = 'walking_'
walking_df_full = walking_df_full.rename(columns={col: prefix_2 + col for col in cols_to_add})

walking_df_full.head()
```

```python
merged = listings_copy_df.merge(driving_df_full, on="property_id", how="left")
merged2 = merged.merge(walking_df_full, on="property_id", how="left")
```

```python
df = pd.read_csv("../data/curated/rent_features/cleaned_listings_sampled.csv")

cols_to_drop = (set(merged2.columns) - set(df.columns))

merged2 = merged2.drop(columns=cols_to_drop)

mask = merged2.iloc[:, -6:].isna().any(axis=1)
merged2 = merged2.loc[~mask].copy()

merged2 = merged2.dropna(subset=['coordinates'])

merged2
```
