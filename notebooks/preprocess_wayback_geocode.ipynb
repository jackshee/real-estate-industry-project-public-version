```python
# reload all module before executing code
%load_ext autoreload
%autoreload 2

import pandas as pd
import os 

# Import the GeoUtils class from utils/geo.py
from utils.geo import GeoUtils
from dotenv import load_dotenv

load_dotenv()
APIKEY1 = os.getenv('ORS_API_KEY1')
APIKEY2 = os.getenv('ORS_API_KEY2')
APIKEY3 = os.getenv('ORS_API_KEY3')
APIKEY4 = os.getenv('ORS_API_KEY4')
# Initialize GeoUtils 
geoutils = GeoUtils(ors_api_key=APIKEY3)
```

```python
df_tmp = pd.read_csv('../data/raw/missing_geo/rental_listings_missing_coordinates.csv')

df_tmp.info()
```

```python
# process first 500 rows of df 
df_tmp = df_tmp.head(500)
print(df_tmp.shape)
# apply geocoding using geoutils.geocode_ors_with_delay
df_tmp['coordinates'] = df_tmp['full_address'].apply(geoutils.geocode_ors_with_delay)
```

```python
df_tmp.head()
```

```python
# We read in the cleaned_listings.csv file that needs to be imputed with the API result
df = pd.read_csv('../data/curated/rent_features/cleaned_listings.csv')

print("Number of missing coordinates before imputation: ", df['coordinates'].isnull().sum())

coord_mapping = df_tmp.set_index('property_id')['coordinates'].to_dict()

mask = df['coordinates'].isna()

df.loc[mask, 'coordinates'] = df.loc[mask, 'property_id'].map(coord_mapping)

print("Number of missing coordinates after imputation: ", df['coordinates'].isnull().sum())

# save the imputed results
df.to_csv('../data/curated/rent_features/cleaned_listings.csv', index=False)
```

```python
# Now that we've geocoded the first 500 rows, we will update the rental_listings_missing_coordinates.csv file
df_tmp = pd.read_csv('../data/raw/missing_geo/rental_listings_missing_coordinates.csv')
# drop the first 500 rows
df_tmp = df_tmp.iloc[500:]
# save the updated file
df_tmp.to_csv('../data/raw/missing_geo/rental_listings_missing_coordinates.csv', index=False)
```
